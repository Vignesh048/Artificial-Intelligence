### Breadth First Search

```python
def BreadthFirstSearch(Tree,s,g):
    f = [s]
    l = []
    
    #print('l',l)
    #print('f',f,'\n\n')
    
    while l == [] or  l[-1] != g :  # (l==[] at the start of loop, 
                                    # l[-1] != g ->loop continues until reaching goal node)
            
        if f[0] not in l:
            l.append(f[0])# [ .....s] adding the first element of f to the l at the end
            #print('l',l)
            f = f + Tree[f[0]][0]  # f = [s,a,.. @,...] adding children nodes(@,....) 
            #of first element(a) of f at the end of f 
            #print('f',f)
            f.remove(f[0]) #  [a......] removing the first element(s) of f which is added to l 
            #print('f',f,'\n\n')
        else:
            f.remove(f[0])
              #print('f',f,'\n\n')
    return l
```

### Depth First Search

```python
def DepthFirstSearch(Tree,s,g):
    f = [s]
    l = []

    #print('l',l)
    #print('f',f,'\n')
    while l == [] or  l[-1] != g :  # (l==[] at the start of loop, 
                                    # l[-1] != g ->loop continues until reaching goal node)
            
            if f[0] not in l:
                l.append(f[0])# [ .....s] adding the first element of f to the l at the end
                #print('l',l)
                f = Tree[f[0]][0] + f  # f = [@,...,s,a,.. ] adding children nodes(@,...)
                #of  previous first element(a) of f at the start of l 
                #print('f',f)
                f.remove(l[-1])#  [a......] removing the last element(s) of l 
                #print('f',f,'\n\n')
            else:
                f.remove(f[0])
                #print('f',f,'\n\n')
    return l
```

### Uniform Cost Search

```python
def UniformCostSearch(Tree,s,g):
    f = [s]
    l = []
    c = [0]
    
    #print('l',l)
    #print('f',f)
    #print('c',c,'\n\n')

    while l == [] or  l[-1] != g :  # (l==[] at the start of loop, 
                                    # l[-1] != g ->loop continues until reaching goal node)
            
            e = f[c.index(min(c))]# Least cost node
            if e not in l:
                l.append(e) # Adding Least cost node to the end of final list l
                #print('l',l)
                f = f  + Tree[e][0] #Adding children nodes of least cost node to first least l
                #print('f',f)
                c = c  + [x + min(c) for x in Tree[e][1]] # Adding cost of those children nodes
                #print('c',c)
                f.pop(c.index(min(c)))  # removing least code node from first list f
                #print('f',f)
                c.remove(min(c)) # removing cost of least cost node from cost list c
                #print('c',c,'\n\n')
            else:
                c.pop(f.index(e))
                f.remove(e)
                #print('f',f)
                #print('c',c,'\n\n')
        
    return l
```
